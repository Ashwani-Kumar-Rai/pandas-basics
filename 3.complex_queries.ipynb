```python
# NOTE : Concept of loc and iloc is important  , yad krlo loc,iloc hi use krna hai values set krne k liye
import numpy as np
import pandas as pd
newdf = pd.DataFrame(np.random.rand(334,5),index=np.arange(334)) 
newdf.loc[(newdf[0]<0.3)]
```

```python
newdf.loc[(newdf[0]<0.3) & (newdf[3]>0.1)]
```

```python
# Note : Name use krna hai to loc nhi to iloc 

# iloc

newdf.iloc[0,4] # row ka 0 ,column ka 4 
```

```python
newdf.iloc[[0,4],[1,2]]
```

```python
newdf.head(3)
```

```python
# To drop a particular row 
newdf.drop([0])
```

```python
# To drop a particular column 

newdf.drop([0],axis=1)
```

```python
# delete column 1 and 4 
newdf.drop([1,4],axis=1) # axis = 1 means column
# newdf.drop([1,4],axis=1,inplace=True) # inplace=True means original ko hi modify krdo 
```

```python

newdf.head(3)
```

```python
# To reset the index number , but ye ek column aur add krdeta hai index name ka 
newdf.reset_index()
```

```python
# index wala column delete hojayega
newdf.reset_index(drop=True,inplace=True)
```

```python
# column 1 k null kidahar hai ?
newdf[1].isnull()
```

```python
# column 1 ki values ko none krna hai 
newdf[1] = None
newdf
```

```python
# setting a colum's value using loc
newdf.loc[:,[1]]=55
newdf
```
